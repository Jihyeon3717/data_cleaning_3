```python
trip_data['payment_method'] = trip_data['payment_method'].replace(['Credit Card', 'Debit Card'], 'Card')
```

```python
print(trip_data['payment_method'].value_counts())
```

```python
# datetime 형식으로 변환
trip_data['tpep_pickup_datetime'] = pd.to_datetime(trip_data['tpep_pickup_datetime'])
trip_data['tpep_dropoff_datetime'] = pd.to_datetime(trip_data['tpep_dropoff_datetime'])

# 주행 시간 계산 (초 단위로)
trip_data['duration'] = (trip_data['tpep_dropoff_datetime'] - trip_data['tpep_pickup_datetime']).dt.total_seconds() / 60

# "몇 분 몇 초" 형식으로 변환
trip_data['duration_formatted'] = trip_data['duration'].apply(lambda x : f"{int(x // 60)}분 {int(x % 60)}초")

# 주행 시간 확인 (분 단위)
print(trip_data[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'duration','duration_formatted']])
```
